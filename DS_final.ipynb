## Task 7 - Statistics

#### Task 7.1 (2 points):

Which type of t-test should be used to determine if there is a statistical difference between the number of strikes (`STR_total`) when there was\wasn't a warning (`WARNED=Yes\No`)

```python
from scipy import stats
import numpy as np
```

```python
group_with_warning = all_strikes.loc[all_strikes['WARNED'] == 'Yes'].select_dtypes(include=['number'])
group_without_warning = all_strikes[all_strikes['WARNED'] == 'No'].select_dtypes(include=['number'])
t_statistic, p_value = stats.ttest_ind(group_with_warning, group_without_warning)
alpha = 0.05

t = len(group_with_warning)+len(group_without_warning)-2

# Calculate the critical t-value
# ppf is used to find the critical t-value for a two-tailed test
critical_t = stats.t.ppf(1 - alpha/2, t)

# Print the results
print("T-value:", t_statistic)
print("P-Value:", p_value)
print("Critical t-value:", critical_t)
print('With T-value')
```
